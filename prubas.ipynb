```python
import sqlite3 as sqlite
def crear_tablas():
    """Crea base de datos y tabla

    Returns:None
    """
    conn = sqlite.connect('database.db') 
    conn.execute('''CREATE TABLE IF NOT EXISTS usuarios 
    (email TEXT PRIMARY KEY NOT NULL, nombre TEXT NOT NULL, password TEXT NULL)''')
    conn.execute('''DROP TABLE spotify''')
    conn.execute('''CREATE TABLE IF NOT EXISTS spotify 
    (id TEXT NOT NULL, name TEXT NOT NULL, artists TEXT NOT NULL, danceability FLOAT NOT NULL, 
    energy FLOAT NOT NULL, loudness FLOAT NOT NULL, speechiness FLOAT NOT NULL, acousticness FLOAT NOT NULL, 
    instrumentalness FLOAT NOT NULL, liveness FLOAT NOT NULL, valence FLOAT NOT NULL, tempo FLOAT NOT NULL,
    duration_ms INT NOT NULL, ranking INT NOT NULL, ranking_5 TEXT NOT NULL, cluster INT NOT NULL)''') 
    conn.close() 
    return None
```

```python
crear_tablas()
```

```python
def insertar_csv_spotify():
    import csv
    path = os.path.join('datos','spotify_cl.csv')
    try: # Lo intentamos
        con = sqlite.connect("database.db")
        cur = con.cursor()
        with open(path, 'r') as x:
            csv_reader = csv.reader(x)
            next(csv_reader) # Fila columnas
            for fila in csv_reader:
                cur.execute("INSERT INTO spotify (id, name, artists, danceability, energy, loudness, speechiness, acousticness, instrumentalness, liveness, valence, tempo, duration_ms, ranking,ranking_5, cluster) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)",
                (fila[0], fila[1], fila[2], fila[3], fila[4], fila[5], fila[6], fila[7], fila[8], fila[9], fila[10], fila[11], fila[12], fila[13], fila[14],fila[15]))
        con.commit()
    finally: # Pase lo que pase, cerramos la conexión
        con.close()
```

```python
insertar_csv_spotify()
```

```python
def numericas(df):
    """Filtra las columnas numéricas de un DataFrame y devuelve un nuevo DataFrame
    que contiene solo esas columnas."""
    cols_numericas = df.select_dtypes(include=np.number).columns.tolist()
    cols_numericas = cols_numericas[:-2]# quitar cluster y ranking que no interesa mucho verlo 
    df_numericas = df[cols_numericas]
    
    return df_numericas
```

```python
def graficos_point(df, columna):
    """Esta función devuelve un gráfico de barras del dataframe y la columna indicadas

    Args:
        df (dataframe): dataframe
        columna (string): columna del dataframe (columna tipo string)

    Returns:
        plot: gráfico de barras
    """
    fig, ax = plt.subplots(figsize=(15,20))

    dataframe = pd.DataFrame(df[columna].value_counts())
    dataframe = dataframe.reset_index()
    ax.bar(data=dataframe, x='index', height=columna, color='salmon', alpha=0.8)
    ax.set_title(columna)
    ax.grid(which='both', linestyle='dashed', axis='y')
    ax.set_axisbelow(True)
    ax.set_xlabel(columna)
    ax.set_ylabel('número de reservas')
    props = {"rotation" : 90}
    plt.setp(ax.get_xticklabels(), **props)

    return fig
```

```python
def graficos_hist(columna):
    """Esta función devuelve un histograma de densidad del la columna indicada

    Args:
        columna (str): columna del dataframe (columna tipo numérica)

    Returns:
        plot: histograma de densidad
    """
    fig = sns.distplot(datos[columna], hist=True, kde=True, color='green').set(title=columna)

    return fig
```
